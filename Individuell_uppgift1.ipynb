## Översikt

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Nyckeltal
#### Ålder

```python
df = pd.read_csv('health_study_dataset.csv')

#  Ålder nyckeltal
def age_key_metrics(df):
    age_mean = df['age'].mean()
    age_median = df['age'].median()
    age_min = df['age'].min()
    age_max = df['age'].max()
    return age_mean, age_median, age_min, age_max

age_mean, age_median, age_min, age_max = age_key_metrics(df)

summary_data = [
    ["Medelålder", round(age_mean, 1)],
    ["Medianålder", round(age_median, 1)],
    ["Yngsta deltagare", age_min],
    ["Äldsta deltagare", age_max]
]

summary_df = pd.DataFrame(summary_data, columns=["Nyckeltal", "Ålder"])

# Visar tabellen
summary_df
```

### Visualisering

#### Deltagarnas åldersdistribution

```python
fig, ax =plt.subplots()
ax.hist(df["age"], bins=10)
ax.set_title("Åldersdistribution")
ax.set_xlabel("Ålder")
ax.set_ylabel("Antal deltagare")
ax.grid(True, axis='y')
plt.tight_layout()
```

#### Viktdistribution

```python
fig, ax = plt.subplots(figsize=(8,6))
df.boxplot(column = 'weight', by = 'sex', ax = ax, vert = False)
ax.set_title("Viktdistribution efter kön")
ax.set_xlabel("Vikt (kg)")
ax.set_ylabel("Kön")
plt.suptitle("")
plt.tight_layout()
```

#### BMI och blodtryck
Kliniska studier har visat på ett samband mellan högt BMI och högt blodtryck.

```python
# https://pmc.ncbi.nlm.nih.gov/articles/PMC6316192/#:~:text=We%20found%20a%20gradient%20of,of%20other%20clinical%20risk%20factors.
# https://www.capio.se/sjukdomar-och-besvar/levnadsvanor/BMI/

BMI = df['weight'] / (df['height'] / 100) ** 2        # Beh[ll BMI som en temporär variabel för plotting 
fig, ax = plt.subplots(figsize=(8,6))

x = BMI
y = df['systolic_bp']

ax.scatter(x, y, alpha= 0.7)
ax.set_title("Eventuellt samband mellan BMI och blodtryck bland deltagarna")
ax.set_xlabel("BMI")
ax.set_ylabel("Blodtryck (mmHg)")

plt.tight_layout()
plt.show()

r = BMI.corr(df['systolic_bp'])
print(f"Korrelationskoefficient (r) mellan BMI och blodtryck (mmHg): {r:.2f}")
```
